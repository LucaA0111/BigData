#### ***Task 2: Produttori di Batterie***

[***Produttori di batterie excel***](https://docs.google.com/spreadsheets/d/1COCWG_WU3gC-dUSjDs-iRQ-4SbtpAJZ5wgHIaf2Sqfo/edit?usp=sharing)

Il codice carica inizialmente i dati dei due produttori A e B di batterie, da una repository di GitHub utilizzando la libreria "pandas". Calcola poi la **media, la deviazione standard e le distribuzioni normalizzate per entrambi i produttori**. Visualizza infine i risultati ottenuti e i crea grafici per mostrare le distribuzioni normalizzate dei dati.

Possiamo notare che osservando solo la media delle batterie dei due produttori, questi hanno lo stesso comportamento. Se invece osserviamo media e deviazione standard insieme, possiamo notare che vi è un comportamento diverso tra le batterie dei due produttori. Infatti confrontando la deviazione standard dei due produttori notiamo che quella del **produttore B è più bassa**, ciò significa che i valori del dataset sono concentrati di più intorno alla media e questo ci porta alla conclusione che il produttore B utilizza la strategia dell’obsolescenza programmata.

```python
import numpy as np
from scipy.stats import norm
import pandas as pd
import matplotlib.pyplot as plt

# Importiamo i dati da GitHub
url = "https://raw.githubusercontent.com/LucaA0111/BigData/main/Statistica/ProduttoriBatterie.csv"
df = pd.read_csv(url)

# Calcoliamo Media, Deviazione Standard e Distribuzione Normalizzata del produttore A e mandiamo in output le informazioni ottenute
Media_prodA = df['PRODUTTORE A'].mean()
Stan_devA = df['PRODUTTORE A'].std()
Distr_norA = np.zeros(len(df['PRODUTTORE A']))
for i in range(len(df['PRODUTTORE A'])):
  Distr_norA[i] =norm.pdf(df.iloc[i,0], loc = Media_prodA, scale = Stan_devA)

# Calcoliamo Media, Deviazione Standard e Distribuzione Normalizzata del produttore B e mandiamo in output le informazioni ottenute
Media_prodB = df['PRODUTTORE B'].mean()
Stan_devB = df['PRODUTTORE B'].std()
Distr_norB = np.zeros(len(df['PRODUTTORE B']))
for i in range(len(df['PRODUTTORE B'])):
  Distr_norB[i] = norm.pdf(df.iloc[i,1], loc = Media_prodB, scale = Stan_devB)

# Stampiamo i risultati ottenuti
print("Produttore A\n")
print("Media batterie produttore A: " + str(Media_prodA) + "\n")
print("Deviazione standard A: " + str(Stan_devA) + "\n")
print("Distribuzione normalizzata di A: " + str(Distr_norA) + "\n")
print("----------------------------------------\nProduttore B\n")
print("Media batterie produttore B: " + str(Media_prodB) + "\n")
print("Deviazione standard B: " + str(Stan_devB) + "\n")
print("Distribuzione normalizzata di B: " + str(Distr_norB) + "\n")


# Creiamo i grafici
fig, ax = plt.subplots(1, 1, figsize = (10,5))
plt.plot(df['PRODUTTORE A'], Distr_norA)
plt.title('PRODUTTORE A')
plt.xlabel('Durata Batterie')
plt.ylabel('Normalizzazione')

fig, ax = plt.subplots(1, 1, figsize = (10,5))
plt.plot(df['PRODUTTORE B'], Distr_norB)
plt.title('PRODUTTORE B')
plt.xlabel('Durata Batterie')
plt.ylabel('Normalizzazione')

plt.show()
```

#### ***Task 2: Pearson vs Spearman***

[***Pearson vs Spearman EXCEL***](https://docs.google.com/spreadsheets/d/1nl5W9YB16CDCwW5XdaGPEDJSqUdm1JYbTCDvW5HlQi4/edit#gid=1371144962)

A partire dal dataset della Tabella di Mollier fornito, il codice crea due liste di dati: T e Vi. Calcola quindi **la correlazione di Pearson ed il coefficiente di Spearman** tra le due liste, grazie alle funzioni offerte dalla libreria scipy.stats. Stampa poi i valori di T, Vi, la correlazione di Pearson e il coefficiente di Spearman. Infine, crea un grafico di regressione lineare utilizzando seaborn e matplotlib, rappresentando T sull'asse x e Vi sull'asse y. Questo grafico viene, come ultimo, visualizzato a schermo da cui si può notare la presenza di un outlier.

Guardando le due correlazioni possimo notare che la correlazione di Spearman lavora meglio con le relazioni non lineari mentre la correlazione di Pearson lavora bene con i predittori lineari.

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import spearmanr
import seaborn as sb

T = [0, 5, 10, 15, 20, 30, 40, 60, 80, 100, 150, 200, 250, 300, 374]
Vi = [0.0179, 0.0179, 0.0179, 0.0179, 0.0179, 0.018, 0.018, 0.0183, 0.0185, 0.0187, 0.0196, 0.207, 0.0224, 0.0252, 0.057 ]

# Calcoliamo pearson e spearman e li mandiamo in output
corre, _ = pearsonr (T, Vi)
corr, p_value = spearmanr(T, Vi)
print("T: " + str(T) + "\n")
print("Vi: " + str(Vi) + "\n")
print('Correlazione di Pearson: %.3f' % corre + "\n")
print("Coefficiente di Spearman: ", corr)
print("\n")

# Creiamo il grafico
sb.regplot(x = T, y = Vi, ci = None)
plt.xlabel('T')
plt.ylabel('Vi')
plt.title('')

# Mandiamo in output il grafico ottenuto
plt.show()
```

#### ***Task 3: Traffico a Milano***

[***TRAFFICO A MILANO EXCEL***](https://docs.google.com/spreadsheets/d/1nl5W9YB16CDCwW5XdaGPEDJSqUdm1JYbTCDvW5HlQi4/edit#gid=216529572)

Il codice fornito utilizza diverse librerie come NumPy, matplotlib e pandas per eseguire operazioni di analisi dei dati e visualizzazione grafica su un dataset che rapprenta **l'andamento del traffico milanese**. Una volta importati i dati (consultabili nel link google drive sopra mostrato), vengono effettuati calcoli statistici per ottenere **l'autocorrelazione** (grazie all'applicazione della funzione np.corrcoef() di NumPy) ed i risultati vengono stampati a schermo. Infine, vengono creati due grafici distinti (utilizzando la funzione plt.subplots() di matplotlib) per rappresentare i dati in modo visivo all'utente: il primo raffigura l'autocorrelazione, mentre il secondo descrive il traffico.

Correlando i dati dal dataset con un suo shift, ossia utilizando l'**autocorrelazione**, possiamo notare una **ciclicità** nei valori. Questa ciclicità è evidente nel grafico della Autocorrelazione dove ogni **7 giorni** abbiamo un picco.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importiamo i dati da GitHub
url ="https://raw.githubusercontent.com/LucaA0111/BigData/main/Correlazione/Traffico.csv"
df = pd.read_csv(url)

# Prendiamo i primi sette valori del traffico su cui calcolare l'autocorrelazione e li trasformiamo in array nunmpy
mini_dataset = df.iloc[0:7, 2].to_numpy()
shift = df.iloc[0:21,5].to_numpy().astype(int)

# Creaimo l'array per i valori della correlazione
correlazione = []

# Calcoliamo l'autocorrelazione
for i in range(len(shift)):
  corr = np.corrcoef(mini_dataset, df.iloc[shift[i]:shift[i]+7, 2])[0, 1]
  correlazione.append(corr)

# Creiamo il grafico dell'autocorrelazione
fig, ax = plt.subplots(1, 1, figsize = (10,5))
plt.plot(shift, correlazione)
plt.grid(True)
plt.xticks(np.arange(np.min(shift), np.max(shift)+1, 1))
plt.title('Autocorrelazione')
plt.xlabel('Shift Giorni')
plt.ylabel('Autocorrelazione')

# Creiamo il grafico del traffico
fig, axs = plt.subplots(1, 1, figsize = (10,5))
plt.plot(df['Giorno_1'], df['Traffico'])
plt.grid(True)
plt.xticks(np.arange(df['Giorno_1'].min(),df['Giorno_1'].max()+1,1))
plt.title('Traffico a Milano')
plt.xlabel('Giorni')
plt.ylabel('Indice di Traffico')

# Mandiamo in output i grafici ottenuti
plt.show()
```



▶ ***CALCOLO DEI COEFFICIENTI DI PEARSON E SPEARMAN***

Il codice calcola la **correlazione tra Bitcoin e le altre crypto** del dataset, utilizzando sia il coefficiente di correlazione di Pearson che il coefficiente di correlazione di Spearman. Calcola inoltre il valore del **p-value** utilizzando il test t di Student per valutare la significatività della correlazione. Ottiene poi il valore di **R quadro** come il quadrato del coefficiente di correlazione di Spearman e manda in output una tabella che riporta tutti i dati.

In tutti e tre i casi vi è una correlazione ***positiva*** e si può notare che essa è maggione tra Bitcoin ed Ethereum. Questo significa che all'aumentare del valore dei bitcoin anche le altre criptovalute aumenteranno (coesiste quindi il principio di causalità). Vista la presenza di outlier e la non linearità delle relazioni la correlazione migliore da utilizzare è quella di ***Spearman***.

Il **P-value** è uguale a 0, ciò significa che il risultato è **statisticamente significativo**.



```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from scipy import stats
import seaborn as sb

# Importiamo i dati da GitHub
url = "https://raw.githubusercontent.com/LucaA0111/BigData/main/Correlazione/Crypto.csv"
df = pd.read_csv(url)

# Calcoliamo la correlazione con Pearson
corre_BitEth, _ = pearsonr (df['Bitcoin'], df['Ethereum']) #Bit-Eth
corre_BitDog, _ = pearsonr (df['Bitcoin'], df['Dogecoin']) #Bit-Dog
corre_BitLite, _ = pearsonr (df['Bitcoin'], df['Litecoin']) #Bit-Lite

# Calcoliamo la correlazione con Spearman
corr_BitEth, p_value = spearmanr(df['Bitcoin'], df['Ethereum']) #Bit-Eth
corr_BitDog, p_value = spearmanr(df['Bitcoin'], df['Dogecoin']) #Bit-Dog
corr_BitLite, p_value = spearmanr(df['Bitcoin'], df['Litecoin']) #Bit-Lite

# Calcoliamo il P-value
t_statistic, p_valueBE = stats.ttest_ind(df['Bitcoin'], df['Ethereum']) #Bit-Eth
t_statistic, p_valueBD = stats.ttest_ind(df['Bitcoin'], df['Dogecoin']) #Bit-Dog
t_statistic, p_valueBL = stats.ttest_ind(df['Bitcoin'], df['Litecoin']) #Bit-Lite

# Calcoliamo il valore di r quadro
r_quadro_BitEth= corr_BitEth ** 2
r_quadro_BitDog= corr_BitDog ** 2
r_quadro_BitLite= corr_BitLite ** 2

# Mandiamo in output la Tabella
tabella = {'Correlazione|': ['Pearson |', 'Spearman |', 'P-value |', 'r quadro |'],
           'Bitcoin - Ethereum|': ['%.3f' % corre_BitEth + "|", '%.3f' % corr_BitEth + "|", '%.3f' % p_valueBE + "|", '%.3f' % r_quadro_BitEth + "|"],
        'Bitcoin - Dogecoin|': ['%.3f' % corre_BitDog + "|", '%.3f' % corr_BitDog + "|", '%.3f' % p_valueBD + "|", '%.3f' % r_quadro_BitDog + "|"],
        'Bitcoin - Litecoin|': ['%.3f' % corre_BitLite + "|", '%.3f' % corr_BitLite + "|", '%.3f' % p_valueBL + "|", '%.3f' % r_quadro_BitLite + "|"] }
tb = pd.DataFrame(tabella)
print(tb)
print('\n')
print("r quadro BitEth:" + str(r_quadro_BitEth))

# Creiamo i grafici
fig, axs = plt.subplots(1, 1, figsize = (10,5))
sb.regplot(x = df['Bitcoin'], y = df['Ethereum'], ci = None)
plt.xlabel('Bitcoin')
plt.ylabel('Ethereum')
plt.title('Correlazione Bitcoin - Ethereum')

fig, axs = plt.subplots(1, 1, figsize = (10,5))
sb.regplot(x = df['Bitcoin'], y = df['Dogecoin'], ci = None)
plt.xlabel('Bitcoin')
plt.ylabel('Dogecoin')
plt.title('Correlazione Bitcoin - Dogecoin')

fig, axs = plt.subplots(1, 1, figsize = (10,5))
sb.regplot(x = df['Bitcoin'], y = df['Litecoin'], ci = None)
plt.xlabel('Bitcoin')
plt.ylabel('Litecoin')
plt.title('Correlazione Bitcoin - Litecoin')

# Mandiamo in output i grafici ottenuti
plt.show()
```

In questo codice vengono importati i dati presenti in una pagina web (url:https://it.wikipedia.org/wiki/Classifica_delle_Università_Italiane_per_numero_di_studenti) per ottenere **informazioni da una tabella**. Viene utilizzata la libreria **BeautifulSoup** per analizzare il **contenuto HTML** della pagina. Dopo aver eseguito una richiesta HTTP per ottenere il contenuto della pagina, si individua la tabella all'interno del documento HTML utilizzando i metodi di BeautifulSoup. Successivamente, si estraggono i dati di interesse dalla tabella, come il **rango, il nome dell'università, la città, regione ed il numero di studenti**. Infine, i valori estratti vengono stampati in output.

```python
from bs4 import BeautifulSoup
import requests

# Definiamo l'URL della pagina contenente la tabella
url = 'https://it.wikipedia.org/wiki/Classifica_delle_Università_Italiane_per_numero_di_studenti'

# Eseguiamo una richiesta HTTP per ottenere il contenuto della pagina
response = requests.get(url)

# Analizziamo il contenuto della pagina utilizzando la libreria BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Individuiamo la tabella all'interno del documento HTML e tutte le sue righe
table = soup.find('table', class_='wikitable sortable')
rows = table.find_all('tr')

# Mandiamo in output i valori di interesse estratti man mano dalla tabella
for row in rows[1:]:
    cells = row.find_all('td')
    rank = cells[0].get_text().strip()
    name = cells[1].get_text().strip()
    students = cells[2].get_text().strip()
    city = cells[3].get_text().strip()
    num_stud = cells[4].get_text().strip()
    print(rank, name, students, city, num_stud)


```

#### ***Task 2: ELO Ranking***

[***ELO Ranking EXCEL***](https://docs.google.com/spreadsheets/d/1oDmMV0kbTNX1z1jQ5G6HEZ9DAQqnVDTtiBSSqlYxoKQ/edit#gid=0)

Il codice implementa una funzione per il **calcolo automatico del punteggio ELO** nel gioco degli scacchi e crea una **simulazione di calcolo** del punteggio. Vengono definiti i parametri per il calcolo del nuovo Elo: il **coefficiente correttivo di variazione** Elo (**K**), una lista per memorizzare i punteggi **Ea** (**Expected score**) e **Sa** (**Score actual**), una lista "Ris" per memorizzare i risultati delle partite (**Vittoria, Pareggio, Sconfitta**) ed una (Elo_avversario) per memorizzare i punteggi Elo degli avversari. Viene richiesto l'inserimento del punteggio Ea di partenza, l'ELO personale e il numero di avversari. Successivamente, vengono inseriti i punteggi Elo degli avversari e i risultati delle partite. Il programma calcola l'**aspettativa di successo** EA per ciascun avversario utilizzando la formula EA = 1 * P_A_vs_B + (-1) * (1 - P_A_vs_B), dove P_A_vs_B rappresenta la probabilità che il giocatore corrente vinca contro l'avversario, calcolata sulla base della differenza dei loro punteggi Elo.  Il nuovo punteggio Elo viene calcolato utilizzando l'apposita formula  ovvero nuovo_elo = tuo_elo + K * (tot_sa - sum(Ea) / n_avversari) e mandato in output. Infine, vengono visualizzate  le statistiche delle partite in un grafico a barre.

```python
import matplotlib.pyplot as plt

# Definiamo i parametri per il calcolo del nuovo Elo
K = 32  # Coefficiente correttivo di variazione Elo
Ea = []
Sa = []
Ris = []
Elo_avversario = []

# Chiediamo in input l'inserimento del proprio EA, ELO ed il numero di avversari con cui si ha giocato
Ea_partenza = float(input("Inserisci il tuo EA: "))
tuo_elo = int(input("Inserisci il tuo ELO: "))
n_avversari = int(input("Inserisci il numero di avversari con cui hai giocato: "))

# Chiediamo in input l'inserimento dell'ELO e il risultato di ogni partita degli avversari
for i in range(n_avversari):
    elo = int(input("Inserisci l'ELO ranking dell'avversario {}: ".format(i+1)))
    Elo_avversario.append(elo)

for j in range(n_avversari):
    risultato = str(input("Inserisci l'esito della partita con l'avversario {} (V, P, S): V vittoria, P pareggio, S sconfitta ".format(j+1)))
    Ris.append(risultato)

# Calcoliamo il totale dei risultati (1 per la vittoria, 0.5 per il pareggio, 0 per la sconfitta)
for d in range(n_avversari):
    if Ris[d] == "V":
        Sa.append(1)
    elif Ris[d] == "P":
        Sa.append(0.5)
    else:
        Sa.append(0)
tot_sa = sum(Sa)

# Calcoliamo il nuovo valore di Elo
for i in range(n_avversari):
    x = tuo_elo - Elo_avversario[i]
    P_A_vs_B = 1 / (1 + 10 ** (x / 400))
    Ea.append(1 * P_A_vs_B + (-1) * (1 - P_A_vs_B))

nuovo_elo = tuo_elo + K * (tot_sa - sum(Ea) / n_avversari)

# Mandiamo in output il valore del nuovo Elo ottenuto
print("\n" + "NUOVO ELO: " + str(int(nuovo_elo)) + "\n")

# Creiamo il grafico delle statistiche
vittorie = 0
patte = 0
sconfitte = 0
for i in range(n_avversari):
    if Ris[i] == "V":
        vittorie += 1
    elif Ris[i] == "P":
        patte += 1
    else:
        sconfitte += 1

# Creiamo il relativo grafico a barre
plt.xticks([], [])
plt.bar(1, vittorie, color='green', label='Vittorie')
plt.bar(3, patte, color='orange', label='Pareggi')
plt.bar(5, sconfitte, color='red', label='Sconfitte')
plt.title("Statistiche")
plt.legend()

# Mandiamo in output il grafico
plt.show()
```

```python
#!pip install wikipedia-api
import networkx as nx
import wikipediaapi
import matplotlib.pyplot as plt

# Chiediamo all'utente di inserire il nome del personaggio di cui sapere il page ranking
personaggio = input("Inserisci il nome del personaggio di cui vuoi trovare il page ranking: ")

# Creiamo un wikipedia api wrapper
wiki = wikipediaapi.Wikipedia(
        language='it',
        extract_format=wikipediaapi.ExtractFormat.WIKI
)

# Creiamo un grafo vuoto
G = nx.DiGraph()

# Recuperiamo la pagina di Wikipedia per il personaggio scelto
page = wiki.page(personaggio)

# Aggiungiamo i nodi
G.add_nodes_from([page.title])

# Recuperiamo i link interni alla pagina
for link in page.links.items():
    G.add_nodes_from([link[0]])
    G.add_edge(page.title, link[0])

# Calcoliamo il PageRank
pr = nx.pagerank(G)

# Stampiamo il PageRank per ciascun nodo
for node, score in pr.items():
    print(f"Node {node} ha un PageRank di {score}")
```

# ⏩ ***Significatività Statistica***


##### ***Task 1: T TEST***

Il codice esegue un'analisi statistica dei punteggi di **QI** di due gruppi: **maschi e femmine**. Vengono calcolate le medie, le deviazioni standard e altre statistiche dei due gruppi. Viene quindi eseguito un **test T** per confrontare i gruppi e determinare se ci sono differenze significative tra di essi ( il livello di significatività viene scelto dall'utente). Infine, viene generato un grafico per visualizzare la distribuzione dei punteggi di QI dei due gruppi.

```python
import scipy.stats as stats
from scipy.stats import skew
import statistics
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import t

#Dataset dei punteggi del QI dei maschi e delle femmine
QI_maschi = [130, 113, 111, 117, 122, 118,120, 114, 112, 116, 120]
QI_femmine = [119, 126, 130, 121, 121, 117, 139, 130, 120, 114, 120]

# Calcoliamo la media del QI dei maschi e delle femmine
media_maschi = statistics.mean(QI_maschi)
media_femmine = statistics.mean(QI_femmine)

# Calcoliamo la deviazione standard dei punteggi del QI dei maschi e delle femmine
maschi_DevStd = statistics.stdev(QI_maschi)
femmine_DevStd = statistics.stdev(QI_femmine)

#Calcoliamo le mediane dei punteggi del QI dei maschi e delle femmine
mediana_maschi = statistics.median(QI_maschi)
mediana_femmine = statistics.median(QI_femmine)

#Calcoliamo l'assimetria dei punteggi del QI dei maschi e delle femmine
asim_maschi=skew(QI_maschi)
asim_femmine=skew(QI_femmine)

#Distribuzione normale dei punteggi del QI dei maschi e delle femmine
range_valori = np.linspace(min(QI_maschi + QI_femmine) - 10, max(QI_maschi + QI_femmine) + 10, 100)
DistrNorm_maschi = stats.norm.pdf(range_valori, media_maschi, maschi_DevStd)
DistrNorm_femmine = stats.norm.pdf(range_valori, media_femmine, femmine_DevStd)

# Chiediamo all'utente di scegliere il livello di significatività
liv_significatività = float(input('Inserisci il livello di significatività del test: '))

# Calcoliamo i gradi di libertà
gradi_Libertà = len(QI_maschi) + len(QI_femmine) - 2

# Calcoliamo il t-statistico dei punteggi del QI dei maschi e delle femmine
t_statistico = abs(media_maschi - media_femmine) / math.sqrt((maschi_DevStd**2 / len(QI_maschi)) + (femmine_DevStd**2/ len(QI_femmine)))

# Calcoliamo il valore v che la statistica t deve superare
alfa= 1-liv_significatività
v = t.ppf(1-alfa/2, gradi_Libertà)
# Mandiamo in output le informazioni ottenute
print("Numerosità Dataset dei maschi: "+ str(len(QI_maschi)))
print("Numerosità Dataset delle femmine: "+ str(len(QI_femmine)))
print("Media maschi: "+ str(media_maschi))
print ("Media femmine: "+ str(media_femmine))
print("Deviazione standard maschi: "+ str(maschi_DevStd))
print("Deviazione standard femmine: "+ str(femmine_DevStd))
print("Mediana maschi: "+ str(mediana_maschi))
print("Mediana femmine: "+ str(mediana_femmine))
print("L'asimetria del dataset dei maschi: "+ str(asim_maschi))
print("L'asimetria del dataset delle femmine: "+ str(asim_femmine))
print("Livello di significatività scelto dall'utente: "+ str(liv_significatività))
print("Gradi di libertà: "+ str(gradi_Libertà))
print("il t statistico è: "+ str(t_statistico))
print("il valore di v è: "+ str(v))

# Facciamo il confronto tra t_statistico ed il valore v
if t_statistico > v:
    print("L'osservazione è significativa per "+ str(alfa))
else:
    print("L'osservazione NON è significativa")

# Creiamo il grafico della distribuzione normale dei maschi e delle femmine
plt.plot(range_valori, DistrNorm_maschi, label="Maschi", color="blue")
plt.plot(range_valori, DistrNorm_femmine, label="Femmine", color="pink")
plt.xlabel("QI")
plt.ylabel("Densità di probabilità")
plt.title("Distribuzione normale dei punteggi del QI dei maschi e delle femmine")
plt.legend()
plt.show()

```

#### ***Task 2: Kolmogorov-Smirnov Test***


Il programma esegue un **test di Kolmogorov-Smirnov** per determinare se i dati del punteggio del QI per maschi e femmine seguono una distribuzione normale. Vengono calcolati i **valori attesi e reali** ordinando il dataset in maniera crescente, e viene calcolata la **distanza massima tra di essi**. Viene anche confrontato il valore di questa distanza con un **valore critico** (definito in base alla tabella oracolo fissando alfa a 0,05) calcolato in base ad un livello di significatività . Infine, vengono forniti i risultati e viene mostrato un grafico dei valori attesi e reali per entrambi i gruppi.

```python
import math
import statistics
from scipy.stats import norm
import matplotlib.pyplot as plt

# Dataset dei punteggi del QI dei maschi e delle femmine
QI_maschi = [130, 113, 111, 117, 122, 118,120, 114, 112, 116, 120]
QI_femmine = [119, 126, 130, 121, 121, 117, 139, 130, 120, 114, 120]

# Calcoliamo la media del QI dei maschi e delle femmine
media_maschi = statistics.mean(QI_maschi)
media_femmine = statistics.mean(QI_femmine)

# Calcoliamo la deviazione standard dei punteggi del QI dei maschi e delle femmine
maschi_DevStd = statistics.stdev(QI_maschi)
femmine_DevStd = statistics.stdev(QI_femmine)

# Ordiniamo i punteggi del QI dei maschi e delle femmine in maniera crescente
QI_maschi = sorted(QI_maschi)
QI_femmine = sorted(QI_femmine)

# Cumulativo
Cumulativo = []
for i in range(1, len(QI_femmine)+1):
    Cumulativo.append(i)

# Calcoliamo il valore atteso per il dataset delle femmine
Val_Att_Femm = []
for i in range(len(QI_femmine)):
    Val_Att_Femm.append((Cumulativo[i]-1)/len(QI_femmine))

# Calcoliamo il valore reale per il dataset delle femmine
Val_Real_Femm = []
for valore in QI_femmine:
    distribuzione_normale_femm = norm.cdf(valore, media_femmine, femmine_DevStd)
    Val_Real_Femm.append(distribuzione_normale_femm)

# Calcoliamo la distanza massima KS tra i valori attesi e quelli reali per il dataset del QI delle femmine
diff_Femm=[]
for i in range(len(Val_Att_Femm)):
    diff_assoluta = abs(Val_Att_Femm[i] - Val_Real_Femm[i])
    diff_Femm.append(diff_assoluta)
D_femm= max(diff_Femm)

# Calcoliamo il valore atteso per il dataset dei maschi
Val_Att_Maschi = []
for i in range(len(QI_maschi)):
    Val_Att_Maschi.append((Cumulativo[i]-1)/len(QI_maschi))

# Calcoliamo il valore reale per il dataset dei maschi
Val_Real_Maschi = []
for valore_maschi in QI_maschi:
    distribuzione_normale_maschi = norm.cdf(valore_maschi, media_maschi, maschi_DevStd)
    Val_Real_Maschi.append(distribuzione_normale_maschi)

# Calcoliamo la distanza massima KS tra i valori attesi e quelli reali per il dataset del QI dei maschi
diff_Maschi=[]
for i in range(len(Val_Att_Maschi)):
    diff_assoluta_maschi = abs(Val_Att_Maschi[i] - Val_Real_Maschi[i])
    diff_Maschi.append(diff_assoluta_maschi)
D_maschi= max(diff_Maschi)

# Definiamo il valore di c(alfa) in base alla tabella oracolo fissato un alfa
alfa = 0.05  #fissiamo il valore di alfa a 0.05
c_alfa = 0.39122  #considerando il valore di alfa, la tabella oracolo riporta il seguente valore di c(alfa)

# Mandiamo in output le informazioni ottenute del dataset delle FEMMINE
print("Il valore di c(alfa) fissato è: "+ str(c_alfa))
print ("Cumulativo: "+ str(Cumulativo))
print("I VALORI ATTESI per il dataset del QI delle FEMMINE sono: ")
for i in range(len(QI_femmine)):
    print(Val_Att_Femm[i])
print ("I VALORI REALI ottenuti dal dataset del QI delle FEMMINE sono: ")
for i in range(len(QI_femmine)):
    print(Val_Real_Femm[i])
print("La distanza massima KS tra i valori attesi e quelli reali del dataset del QI delle FEMMINE è: "+str(D_femm))

# Facciamo il confronto tra la distanza D max del KS test ed il c(alfa) fissato in base all'oracolo SOLO per il dataset delle FEMMINE
print ("Valutiamo se il dataset delle FEMMINE è realmente una distribuzione normale o meno: ")
Val_Confronto = math.sqrt((len(QI_maschi)+len(QI_femmine))/(len(QI_maschi)*len(QI_femmine)))
if D_femm > (c_alfa*Val_Confronto):
    print("Il dataset delle FEMMINE è realmente una distribuzione normale per un certo livello di alfa: "+ str(alfa))
else:
    print("Il dataset delle FEMMINE NON è realmente una distribuzione normale per un certo livello di alfa: "+ str(alfa))

# Mandiamo in output le informazioni ottenute del dataset dei MASCHI
print("i VALORI ATTESI per il dataset del QI dei MASCHI sono: ")
for i in range(len(QI_maschi)):
    print(Val_Att_Maschi[i])
print ("I VALORI REALI ottenuti dal dataset del QI dei MASCHI sono: ")
for i in range(len(QI_maschi)):
    print(Val_Real_Maschi[i])
print("La distanza massima KS tra i valori attesi e quelli reale del dataset del QI dei MASCHI è: "+str(D_maschi))

# Facciamo il confronto tra la distanza D max del KS test ed il c(alfa) fissato in base all'oracolo SOLO per il dataset dei MASCHI
print ("Valutiamo se il dataset dei MASCHI è realmente una distribuzione normale o meno: ")
if D_maschi > (c_alfa*Val_Confronto):
    print("Il dataset dei MASCHI è realmente una distribuzione normale per un certo livello di alfa: "+ str(alfa))
else:
    print("Il dataset dei MASCHI NON è realmente una distribuzione normale per un certo livello di alfa: "+ str(alfa))

# Facciamo il confronto tra la distanza D max del KS test ed il c(alfa) fissato in base all'oracolo per il dataset delle FEMMINE E DEI MASCHI
print("Confronto tra le due distribuzioni MASCHI-FEMMINE:")
D = abs(D_femm-D_maschi)
if D > (c_alfa*Val_Confronto):
    print("Le due distribuzioni, MASCHI-FEMMINE, differiscono significativamente per un certo livello alfa: "+ str(alfa))
else:
    print("Le due distribuzioni, MASCHI-FEMMINE, NON differiscono significativamente per un certo livello alfa: "+ str(alfa))

# Realizziamo il grafico per il dataset DELLE FEMMINE e DEI MASCHI
fig, axs = plt.subplots(2, 1, figsize=(8, 8))
axs[0].plot(QI_femmine, Val_Att_Femm, label="Valori attesi")
axs[0].plot(QI_femmine, Val_Real_Femm, label="Valori reali")
axs[0].legend(loc="upper left")
axs[0].set_xlabel("Punteggio QI femminile")
axs[0].set_ylabel("Probabilità")
axs[0].set_title("PtoP del QI femminile")
axs[1].plot(QI_maschi, Val_Att_Maschi, label="Valori attesi")
axs[1].plot(QI_maschi, Val_Real_Maschi, label="Valori reali")
axs[1].legend(loc="upper left")
axs[1].set_xlabel("Punteggio QI maschile")
axs[1].set_ylabel("Probabilità")
axs[1].set_title("PtoP del QI maschile")
plt.tight_layout()
plt.show()
```

#### ***Task 3: Test Di Permutazione***

Il codice implementa un **test di permutazione uniforme per generare 1 milione di permutazioni casuali** di un insieme di 4 elementi [1, 2, 3, 4]. Le frequenze di queste permutazioni vengono tracciate e visualizzate graficamente utilizzando la libreria **Matplotlib**. Il codice utilizza anche le librerie **Random, Itertools e Math** per generare permutazioni casuali, ottenere tutte le possibili combinazioni e calcolare il fattoriale.

```python
import matplotlib.pyplot as plt
import random
from itertools import permutations
from math import factorial

# Definiamo il numero di elementi n
n = 4

# Creiamo una lista vuota in cui salveremo permutazioni generate
permutazioni_generate = []

# Creiamo una lista di lunghezza n! inizializzando i suoi elementi a zero,
# per memorizzare tutte le frequenze delle permutazioni che verranno generate
frequenze_permutazioni = [0] * factorial(n)

# Eseguiamo 1 milione di run dell'algo per generare randomicamente le permutazioni degli n elementi
# e ne tracciamo la frequenza
for i in range(1000000):
    lista_n = list(range(1, n+1))  # Creiamo una lista con gli n elementi [1,2,3,4]
    random.shuffle(lista_n)
    permutazione = [str(num) for num in lista_n]  # Convertiamo la combinazione casualmente creata degli
                                                  # n elementi in una stringa salvata in 'permutazione',
                                                  # che rappresenta la permutazione corrente degli n elementi
    if permutazione in permutazioni_generate:  # Verifichiamo se la permutazione è già stata generata
                                               #in precedenza
        indice = permutazioni_generate.index(permutazione)  # Recuperiamo l'indice associato alla permutazione
    else:
        indice = sum((lista_n[i]-1) * factorial(n-i-1) for i in range(n))  # Calcoliamo l'indice da associato
                                                                           # alla nuova permutazione
        permutazioni_generate.append(permutazione)  # Aggiungiamo la permutazione alla lista delle
                                                    # permutazioni generate
    frequenze_permutazioni[indice] += 1  # Incrementiamo la frequenza all'indice trovato

# Creiamo un grafico a barre delle frequenze delle permutazioni
plt.bar(range(factorial(n)), frequenze_permutazioni)

# Inseriamo sull'asse x le etichette delle possibili permutazioni
possibili_permutazioni = [str(permutazione) for permutazione in permutations(range(1, n+1))]
plt.xticks(range(factorial(n)), possibili_permutazioni, rotation=90)

# Nominiamo relativamente gli assi
plt.xlabel('Permutazioni')
plt.ylabel('Frequenze')

# Mostriamo il grafico
plt.show()
```

#### ***Task 2: ROC Curve***

Il codice analizza un insieme di dati e calcola la curva **ROC** insieme all'**AUC**, utilizzate per valutare le prestazioni di un modello di **classificazione** di un vaccino e quindi forniscono informazioni sulla capacità del modello di distinguere tra due classi.  I dati sono rappresentati da due array: uno per i valori negativi e uno per i valori positivi. La curva ROC viene costruita considerando i valori cumulativi di entrambi gli array. Successivamente, vengono calcolati i tassi di **falsi positivi (FPR)** e **veri positivi (TPR)** utilizzando sempre i valori cumulativi.
Infine, viene calcolata l'**AUC**, il valore numerico che misura **l'area sottesa alla curva ROC** e viene definita graficamente la curva ROC utilizzando la libreria **matplotlib**.

In questo caso il classificatore ha una capacità discriminante superiore al caso casuale, più il valore è vicino a 1 migliore sarà la performance del nostro modello.

```python
from sklearn.metrics import auc
from matplotlib.font_manager import ttfFontProperty
import numpy as np
import matplotlib.pyplot as plt

# Definiamo i dati
Negative = np.array([0, 34,74,100,127,153,114,19,6,4,0])
Positive = np.array([0, 2,6,10,12,23,55,75,41,30,15])

# Creiamo la tabella ROC
NegativeC = Negative.cumsum() #Cumulativa per Negative
PositiveC = Positive.cumsum() #Cumulativa per Positive

print("Valori comulativi per i negativi: \n" + str(NegativeC))
print("Valori comulativi per i positivi: \n" + str(PositiveC))

# Calcoliamo FPR
FPR = np.zeros(len(NegativeC))
for i in range(10):
  FPR[i] = np.array([1-NegativeC[i]/NegativeC[10]])

# Calcoliamo TPR
TPR = np.zeros(len(PositiveC))
for i in range(10):
  TPR[i] = 1-PositiveC[i]/PositiveC[10]

#Calcolo AUC
AUC_arr = np.zeros(len(Positive))
for i in range(10):
  AUC_arr[i] = (FPR[i]-FPR[i+1])*TPR[i]

AUC = AUC_arr.sum()

print("\nArea Under the Curve: " + str(AUC))


# Mostriamo il grafico
plt.scatter(FPR, TPR)
plt.title("ROC Curve for the new Nasal Vaccine")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Importimo il dataset da GitHub
df = pd.read_csv("https://raw.githubusercontent.com/LucaA0111/BigData/main/MachineLearning/Bayes.csv")

# Creiamo i dati necessari per l'algoritmo
X = df['Keyword'] + ' ' + df['Lunghezza'] + ' ' + df['Intestazione']
y = df['Valida?']

# Creiamo il vettore di features
vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)

# Suddividiamo il dataset in training set e test set
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)
# La funzione esegue la suddivisione dei dati in un training set e un test set,
# test_size specifica quanti dati assegnare al test set. (20% in questo caso)
# mentre il restante (80%) verrà utilizzato per il training set.
# random_state fornisce un valore per inizializzare il generatore di numeri casuali.

# Creiamo l'istanza del classificatore Naive Bayes
naive_bayes = MultinomialNB()

# Addestriamo il modello utilizzando il training set
naive_bayes.fit(X_train, y_train)

# Effettuiamo le predizioni sul test set
y_pred = naive_bayes.predict(X_test)

# Calcoliamo l'accuratezza del modello
accuracy = accuracy_score(y_test, y_pred)
print("Accuratezza del modello: {:.2f}%".format(accuracy * 100))
```

#### **Task 2: Alberi Decisionali**

Il codice importa le librerie necessarie, carica un dataset da un file CSV (quello dell'esercizio precedente), crea le variabili per l'**algoritmo di apprendimento**, trasforma i dati testuali in features numeriche, **addestra un modello di alberi di decisione** e consente all'utente di inserire valori per prevedere se una certa mail è **spam o non spam** in base a keyword, lunghezza ed intestazione. Nel caso in cui gli inserimenti portino ad un ramo inesistente nell'albero, allora viene mandato in output un opportuno messaggio (Impossibile determinare con precisione se la mail è spam o non spam, perchè il ramo non è presente).

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction.text import TfidfVectorizer

# Importiamo il dataset da GitHub
df = pd.read_csv("https://raw.githubusercontent.com/LucaA0111/BigData/main/MachineLearning/Bayes.csv")

# Creiamo i dati necessari per l'algoritmo
X = df['Keyword'] + ' ' + df['Lunghezza'] + ' ' + df['Intestazione']
y = df['Valida?']

# Inizializziamo il TfidfVectorizer
vectorizer = TfidfVectorizer()

# Trasformiamo i dati testuali in features numeriche
X = vectorizer.fit_transform(X)

# Addestriamo il modello di Alberi di Decisione
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Interazione con l'utente per inserire i valori
print("Inserire dei valori per le seguenti features, in base all'albero decisionale verrà prodotto l'esito")
keyword = input("Inserisci la keyword: ")
lunghezza = input("Inserisci la lunghezza (lunga/media/corta): ")
intestazione = input("L'intestazione è presente? (si/no): ")

# Creazione del dato da prevedere
input_data = keyword + ' ' + lunghezza + ' ' + intestazione

# Trasformiamo il dato testuale in features numeriche
input_data_transformed = vectorizer.transform([input_data])

# Verifica se il modello può fare una previsione
if input_data_transformed.sum() == 0 or not all(feature in vectorizer.vocabulary_ for feature in input_data.split()):
    print("Impossibile determinare con precisione se la mail è spam o non spam.")
else:
    # Prevedi la classe della mail
    prediction = clf.predict(input_data_transformed)[0]

    if prediction == "si":
        print("La mail potenzialmente è spam.")
    elif prediction == "no":
        print("La mail potenzialmente non è spam.")
    else:
        print("Impossibile determinare con precisione se la mail è spam o non spam, perchè il ramo non è presente")
```
